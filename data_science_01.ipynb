### 몬테카를로 시행의 원리

```python
import tensorflow as tf
```

```python
tf.compat.v1.disable_eager_execution()
```

```python
norm = tf.random.normal([100000], mean=0, stddev=2)
with tf.compat.v1.Session() as session:
    plt.hist(norm.eval(), bins=100)
    plt.show()
```

```python
uniform = tf.random.uniform([100000], minval=0, maxval=1, dtype=tf.float32)
with tf.compat.v1.Session() as session:
    # print(uniform.eval())
    plt.hist(uniform.eval(), bins=100)
    plt.show()
```

```python
# 난수로 pi 값 계산

# 반지름이 1인 원 넓이 ; pi
# 원을 외접하는 사각형의 면적 : 4

# 원의 면적/ 사각형의 면적비 = pi/4
# 원 안에 있는 점의 개수 / 사각형 안에 있는 점의 개수(uniform 분포)

# 사각형 안에 점을 10000개 생성 (x 10000개, y 10000개)
# 이 중에서 원 안에 난수가 몇 개 나오는가?

# x^2 + y^2 < 1 : 원 안에 존재
# x^2 + y^2 > 1 : 원 밖에 존재
```

```python
trials = 100
hits = 0
x = tf.random.uniform([1],minval=-1, maxval=1, dtype=tf.float32)
y = tf.random.uniform([1],minval=-1, maxval=1, dtype=tf.float32)
pi = []
sess = tf.compat.v1.Session()

with sess.as_default():
    for i in range(1, trials):
        for j in range(1, trials):
            if x.eval()**2 + y.eval()**2 < 1 :
                hits = hits + 1
                pi.append((4*float(hits) / i)/trials)

plt.plot(pi)
plt.show()

# 난수가 균등분포를 따르는가?
```
