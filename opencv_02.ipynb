### Numpy <-> Image 변환

```python
from PIL import Image
```

```python
image = cv2.imread('./muji.png', cv2.IMREAD_GRAYSCALE)
image_array = np.array(image)
image_array
```

```python
image_array.shape
```

```python
image_from_array = Image.fromarray(image_array)
plt.imshow(image_from_array)
```

```python
# 이미지 저장
cv2.imwrite('./muji_grayscale.png', image)
```

### 이미지 크기 조정 (Resize)

```python
image_50x50 = cv2.resize(image, (50, 50))
image_50x50
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(image_50x50, cmap='gray')
ax[1].set_title('Resized Image')
```

### 이미지 자르기 (Crop)

```python
image_cropped = image[:500, :500]
plt.imshow(image_cropped, cmap = 'gray')
```

```python
image_cropped = image[200:1000, 300:1000]  # [y, x]
plt.imshow(image_cropped, cmap='gray')
```
